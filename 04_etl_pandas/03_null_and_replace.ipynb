Handling Null data

```python
df['street'] = df['street'].fillna('Nao Registrado')
df['email'] = df['email'].fillna('Email Não Cadastrado')
df['number'] = df['number'].fillna('Sem numero')
df['additionals'] = df['additionals'].fillna('Sem Info')
df.head(3)
```

Replace method

```python
df['state'] = df['state'].replace('São Paulo', 'Sao Paulo')
df.head(3)
```

```python
valor_original = 'Sao Paulo'
valor_substituto = 'SP'
df['state'] = df['state'].replace(valor_original, valor_substituto)
df.head(3)
```

```python
df['state'] = df['state'].replace('Acre', 'AC')
df['state'] = df['state'].replace('Rio Grande do Sul', 'RN')
df.head(3)
```

```python
mapeamento = {'Maranhão' : 'Maranhao', 'Piauí' : 'Piaui', 'Goiás' : 'Goias'}
df['state'] = df['state'].replace(mapeamento)
df.head(30)
```
