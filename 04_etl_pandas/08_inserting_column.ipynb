```python
import pandas as pd

df = pd.read_csv(r"D:\engenharia de dados\curso udemy\Arquivos\04.+ETL+Pandas\04. ETL Pandas\arquivos_csv\Clientes.csv", delimiter=',')
df.head()
```

```python
df['Nome Completo'] = df['first_name'] + ' ' + df['last_name']
df.head(2)
```

```python
#Script from last class
df['created_at'] = pd.to_datetime(df['created_at'], format='%d/%m/%Y %H:%M:%S')
df.head()
```

```python
df['ano'] = df['created_at'].dt.year
df.head(2)
```

```python
df['mes'] = df['created_at'].dt.month
df.head(2)
```

```python
df['dia'] = df['created_at'].dt.day
df.head(2)
```

Converting the month to the name of the month

```python
meses = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho',
         'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']
df['Nomemes'] = df['created_at'].dt.month.map(lambda x: meses[x-1])
df.head()
```

```python
df['Hora'] = df['created_at'].dt.hour
df['Minuto'] = df['created_at'].dt.minute
df['Segundo'] = df['created_at'].dt.second
df.head()
```

```python
Ano2018 = df.loc[df['ano'] == 2018]
Ano2018.head()
```

```python
Ano2018mes8 = df.loc[(df['ano'] == 2018) & (df['mes'] == 8)]
Ano2018mes8.head()
```
