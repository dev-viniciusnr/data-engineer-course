```python
import pandas as pd
import os

diretorio = (r"D:\engenharia de dados\curso udemy\Arquivos\Outros+Arquivos\Outros Arquivos")

dados = []

for arquivo in os.listdir(diretorio):
    if arquivo.endswith('.xlsx'):
        caminho_arquivo = os.path.join(diretorio, arquivo)
        df = pd.read_excel(caminho_arquivo)
        dados.append(df)

consolidado = pd.concat(dados)
consolidado.head(5)
```

```python
pasta_destino = r'D:\\engenharia de dados\\curso udemy\\Github\\04_etl_pandas\\consolidar\\'

nome_arquivo = 'VendasGeraldelete.xlsx'

Caminho_Completo = pasta_destino + nome_arquivo 

if os.path.exists(Caminho_Completo):
    os.remove(Caminho_Completo)

consolidado.to_excel(Caminho_Completo, index=False)
```
